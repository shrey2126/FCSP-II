```python
#Directed Graph
G=nx.DiGraph()

G.add_nodes_from([1,2,3,4])
G.add_edges_from([(1,2),(1,3),(1,4),(3,4),(4,3)])

nx.draw_networkx(G)
plt.show()
```

```python
# The following dictionary shows you how 5 people follows eachother on instagram .create a Directed networkx Graph 
#for the given instagram follows
instagram={
    'person1':[0,1,1,0,1],
    'person2':[0,0,1,0,1],
    'person3':[1,1,0,1,1],
    'person4':[1,1,1,0,0],
    'person5':[1,1,0,0,0],
}
people = list(instagram.keys()) 

G=nx.DiGraph()
G.add_nodes_from(people)
for i, person in enumerate(people):
    follows = instagram[person]
    for j, follows_person in enumerate(people):
        if follows[j] == 1:
            G.add_edge(person, follows_person)
            
nx.draw(G, with_labels=True, arrowstyle='->', arrowsize=20,node_size=2100,node_color='pink')
plt.show()

```

```python
#2nd way
import numpy as np
G=nx.DiGraph(np.array(list(instagram.values())))

nx.draw(G, with_labels=True, arrowstyle='->', arrowsize=20,node_size=2100,node_color='pink')
plt.show()
```

```python
#2nd way
import numpy as np
G=nx.DiGraph(np.array(list(instagram.values())))
l={0:'person1',1:'person2',2:'person3',3:'person4',4:'person5'}
nx.draw(G, with_labels=True, arrowstyle='->', arrowsize=20,node_size=2100,node_color='pink',labels=l)
plt.show()
```

# Police Department Incident System;

```python
import pandas as pd
df=pd.read_csv("Police_Department_Incidents_-_Previous_Year__2016_.csv")
x=df['X']
y=df['Y']
category=df['Category']

map1=folium.Map(location=[y.iloc[0],x.iloc[0]],zoom_start=5)

count=0
while count<200:
    folium.Marker(location=[y.iloc[count],x.iloc[count]],tooltip=f"{category[count]}",popup=folium.Popup("""<img src='images.png'/>""",max_width=500)).add_to(map1)
    count+=1

map1
```

```python
m=folium.Map(location=[37.7749,-122.4194],zoom_start=12)

for lat,lon,cat in zip(df['Y'],df['X'],df['Category']):
    folium.Marker(location=[lat,lon],popup=cat).add_to(m)
m.save("Crime.html")
```

# Choropleth  class:

```python
import folium 
import pandas as pd
usa=folium.Map(location=[48,-102],zoom_start=3)
folium.Choropleth(geo_data='us-states.json', 
                  name='choropleth',
                  data=pd.read_csv("US_Unemployment_Oct2012.csv"),
                  columns=['State','Unemployment'],
                  key_on='feature.id',
                  fill_color='YlGnBu',
                  fill_opacity=0.7,
                  line_opacity=0.2,
                  legend_name='Unemployment Sale').add_to(usa)
usa
```
