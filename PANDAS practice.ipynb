```python
#3)Find the number of orders which have ‘Quantity’ less than 3 and which have (either ‘Rating’ greater than 8.5 or ‘Total’
# greater than 600).
no_orders = df[(df['Quantity']<3) & ((df['Rating']>8.5) | (df['Total']>600))]
print(len(no_orders))
```

```python
#4)Find the sum of ‘Total’ purchasing price spent by Member and Normal 'Customer type'.
member_total = df[df['Customer type'] == 'Member']['Total'].sum()
normal_total = df[df['Customer type'] == 'Normal']['Total'].sum()
print("\nTotal purchasing price spent by Customer type:")
print("Member:", member_total)
print("Normal:", normal_total)
```

```python
# 5). Find the percentage of total of ‘gross income’ based on the different ‘Payment’ methods used by customers. (Ewallet, Cash and Credit card)
ewallet_gross = df[df['Payment'] == 'Ewallet']['gross income'].sum()
cash_gross = df[df['Payment'] == 'Cash']['gross income'].sum()
credit_gross = df[df['Payment'] == 'Credit card']['gross income'].sum()
total_gross = df['gross income'].sum()
ewallet_percent = (ewallet_gross / total_gross) * 100
cash_percent = (cash_gross / total_gross) * 100
credit_percent = (credit_gross / total_gross) * 100

print("\nPercentage of total gross income by Payment method:")
print("Ewallet:", ewallet_percent)
print("Cash:", cash_percent)
print("Credit card:", credit_percent)
```

```python
#6)Analyze the purchasing behavior of male and female customers using ‘Gender’ column.
#Find their average purchase prices using ‘Total’ column.
male_avg = df[df['Gender'] == 'Male']['Total'].mean()
female_avg = df[df['Gender'] == 'Female']['Total'].mean()

print("\nAverage purchase prices by Gender:")
print("Male:", male_avg)
print("Female:", female_avg)
```

```python
import matplotlib.pyplot as plt
```

```python
# 7). Create a scatter plot that shows the relationship between total amount spent and rating. (keep ‘+’ marker, with marker size 100 and green color).
plt.figure(figsize=(10, 6))
plt.scatter(df['Total'], df['Rating'], marker='+', s=100, color='green')
plt.title('Relationship between Total Amount Spent and Rating')
plt.xlabel('Total Amount Spent')
plt.ylabel('Rating')
plt.grid(True)
plt.show()
```

```python
# 8). Create a box plot that shows the distribution of ‘Rating’ and ‘Quantity’. And comment about outliers in both columns.
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.boxplot(df['Rating'])
plt.title('Distribution of Rating')
plt.subplot(1, 2, 2)
plt.boxplot(df['Quantity'])
plt.title('Distribution of Quantity')
plt.show()

print("\nOutlier Analysis:")
print("Rating: Outliers are values significantly outside the boxplot's whiskers. If the boxplot has many points beyond the whiskers, it indicates possible outliers. If there are no point beyond the whiskers there are no outliers.")
print("Quantity: Outliers are values significantly outside the boxplot's whiskers. If the boxplot has many points beyond the whiskers, it indicates possible outliers. If there are no point beyond the whiskers there are no outliers.")
```
