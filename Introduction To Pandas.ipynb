```python
def remove_outliers_custom(df, column_name):
    q1=df['horsepower'].quantile(0.25)
    q3=df['horsepower'].quantile(0.75)
    iqr=q3-q1
    upper_limit=q3+1.5*iqr
    lower_limit=q1-1.5*iqr
    df_no_outliers = df[(df[column_name] >= lower_limit) & (df[column_name] <= upper_limit)]
    return df_no_outliers


df_no_outliers = remove_outliers_custom(df, 'horsepower')

# Show the resulting DataFrame without outliers
print("\nDataFrame after removing outliers:")
print(df_no_outliers)
```

```python
import matplotlib.pyplot as plt

# Function to remove outliers
def remove_outliers_custom(df, column_name):
    q1 = df[column_name].quantile(0.25)
    q3 = df[column_name].quantile(0.75)
    iqr = q3 - q1
    upper_limit = q3 + 1.5 * iqr
    lower_limit = q1 - 1.5 * iqr
    df_no_outliers = df[(df[column_name] >= lower_limit) & (df[column_name] <= upper_limit)]
    return df_no_outliers

# Removing outliers from the 'horsepower' column
df_no_outliers = remove_outliers_custom(df, 'horsepower')

# Plotting boxplots before and after removing outliers using only matplotlib
plt.figure(figsize=(12, 6))

# Boxplot before removing outliers
plt.subplot(1, 2, 1)
plt.boxplot(df['horsepower'])
plt.title('Boxplot Before Removing Outliers')

# Boxplot after removing outliers
plt.subplot(1, 2, 2)
plt.boxplot(df_no_outliers['horsepower'])
plt.title('Boxplot After Removing Outliers')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
dFrame=pd.DataFrame({'A':[1,2,np.nan,4,np.nan],'B':[np.nan,6,7,np.nan,np.nan],'C':[10,11,np.nan,np.nan,14]})
dFrame
```

```python
# forward fill-> Up to Down
df1=dFrame.fillna(method="ffill")
df1
```

```python
# backward fill> Down to Up
df1=dFrame.fillna(method="bfill")
df1
```

```python
df2=dFrame.fillna({'A':'AA','B':'BB','C':'CC'})
df2
```

```python
dFrame['A'].fillna(dFrame['A'].mean(),inplace=True)
dFrame['B'].fillna(dFrame['B'].median(),inplace=True)
dFrame['C'].fillna(dFrame['C'].mode()[0],inplace=True)
dFrame
```
