```python
import pandas as pd
import numpy as np
months=range(1,13)
sales=np.array([10,15,20,30,20,50,30,70,80,40,100,10])
expenses=np.array([5,10,15,20,25,30,35,40,45,50,55,60])
plt.plot(months,sales,color="green",label='Sales')
plt.plot(months,expenses,color="red",label='Expenses')
plt.fill_between(months,sales,expenses,where=sales>expenses,alpha=0.25,color="green",interpolate=True)
plt.fill_between(months,sales,expenses,where=sales<expenses,alpha=0.25,color="red",interpolate=True)
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

months = range(1, 13)
sales = np.array([20, 25, 35, 45, 30, 60, 40, 80, 90, 50, 110, 20])
expenses = np.array([10, 18, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70])

plt.plot(months, sales, color="green", label='Sales')
plt.plot(months, expenses, color="red", label='Expenses')
plt.fill_between(months, sales, expenses, where=sales > expenses, alpha=0.25, color="green", interpolate=False)
plt.fill_between(months, sales, expenses, where=sales < expenses, alpha=0.25, color="red", interpolate=False)
plt.legend()
plt.show()
# Making interploate="False" does not color the intersection part
```

# Heat Map:

```python
data=np.random.randint(low=1,high=100,size=(10,10))
print(data)
sns.heatmap(data=data)
plt.show()
```

```python
data=np.random.randint(low=1,high=100,size=(10,10))
print(data)
max=70
min=30
sns.heatmap(data=data,vmin=min,vmax=max)
plt.show()
```

```python
data=np.random.randint(low=1,high=100,size=(10,10))
max=70
min=30
sns.heatmap(data=data,vmin=min,vmax=max,cmap='tab20')
#tab20 & tab10 are both are valid 
plt.show()
```

```python
data=np.random.randint(low=1,high=100,size=(10,10))
max=70
min=30
sns.heatmap(data=data,vmin=min,vmax=max,cmap='tab20',annot=True)
#tab20 & tab10 are both are valid 
plt.show()
```

```python
import requests
from wordcloud import WordCloud,STOPWORDS
url="https://www.gutenberg.org/files/11/11-0.txt"
response=requests.get(url)
text=response.text
print(text)
```

```python
wc=WordCloud(background_color='white',max_words=2000)
wc.generate(text)
plt.imshow(wc)
plt.axis('off')
plt.show()
```

```python
wc=WordCloud(background_color='white',max_words=2000,stopwords=STOPWORDS)
wc.generate(text)
plt.imshow(wc)
plt.axis('off')
plt.show()
```

# Regression Plot:

```python
import matplotlib.pyplot as plt
import seaborn as sns

x = [1, 2, 3, 4, 5, 6]
y = [3, 5, 7, 8, 10, 24]

sns.regplot(x=x, y=y, scatter_kws={'s': 10, 'alpha': 0.7}, line_kws={'color': 'red', 'linewidth': 2})
plt.show()
```
